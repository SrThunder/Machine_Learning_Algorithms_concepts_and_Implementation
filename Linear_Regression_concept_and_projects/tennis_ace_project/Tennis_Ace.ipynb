# Single feature linear regression models

### BreakPointsOpportunities vs. Winnings

```python
x=df[['BreakPointsOpportunities']]
y=df[['Winnings']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.8, test_size = 0.2)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
y_predict = model.predict(x_test)
```

```python
print("Train score:")
print(model.score(x_train, y_train))
print("Test score:")
print(model.score(x_test, y_test))
```

```python
plt.scatter(y_test, y_predict, alpha = 0.4)
plt.xlabel("Actual total winnings")
plt.ylabel("Predicted total winning")
plt.title("Predicted vs Actual total winnings in USD per year")
plt.show()
```
